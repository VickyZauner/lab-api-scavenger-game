## Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

#### 1. Find the secret files.

`````python
# your code here
url = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/'
response_secret = requests.get(url, auth = ('VickyZauner', 'c0dd9105307e296867d3c207f35418098016a927'))
secret_json = response_secret.json()
`````

`````python
secret_json[1]
`````

#### 2.  Sort the filenames ascendingly.

`````python
# your code here
find_secret_asc = np.array([])

for elem in secret_json:
    for key, value in elem.items():
        if key == 'name':
            if value != '.gitignore':
                find_secret_asc = np.append(find_secret_asc, np.array([value]))
            else:
                continue

secret_asc = np.sort(find_secret_asc)
secret_asc
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
for item in secret_asc:
    url = 'https://api.github.com/repos/ironhack-datalabs/scavenger/git/trees/master?recursive=1'
    get_trees = requests.get(url, auth = ('VickyZauner', 'c0dd9105307e296867d3c207f35418098016a927'))
get_trees = get_trees.json()
print(json.dumps(get_trees, indent = 4, sort_keys = True))
`````

`````python
import re

list_scav = []

for elem in get_trees['tree']:
    for keys in elem.items():
        if keys[0] == 'path' and keys[1].endswith('scavengerhunt'):
            list_scav.append(keys[1])

print(len(list_scav))
list_scav_sorted = list(np.sort(list_scav))
print(list_scav_sorted)
print(len(list_scav_sorted))
`````

`````python
# your code here
secret = []
headers = {'Accept': 'application/vnd.github.v3.raw'}

for path in list_scav_sorted:
    url = f'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{path}'
    secret_content = requests.get(url, auth = ('VickyZauner', 'c0dd9105307e296867d3c207f35418098016a927'), headers = headers)
    secret.append(secret_content.text)
`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.
